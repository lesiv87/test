```python
# Создадим функцюю которая откорректирует значение даты в колонке "Дата и время"
def month_correct(x):
  a = x.split('\n')
  dict_month = {'янв': '01',
                'фев': '02',
                'мар': '03',
                'апр': '04',
                'май': '05',
                'июн': '06',
                'июл': '07',
                'авг': '08',
                'сен': '09',
                'окт': '10',
                'ноя': '11',
                'дек': '12'}
  b = a[0].split(' ')
  if b[1] in dict_month:
    b[1] = dict_month[b[1]]
    a[0] = '-'.join(b)

  return ' '.join(a)
```

```python
# С помощью фукции проведем корректировку
df_score['Дата и время'] = df_score['Дата и время'].apply(month_correct)
```
